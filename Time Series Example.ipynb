# Times series
from: https://machinelearningmastery.com/decompose-time-series-data-trend-seasonality/
## TS can be decomposed into, Level, Trend, Seasonality, Noise
Level: average value over time
Trend: growth over time
Seasonality: cyclic behavior
Noise
## TS can be additive or multiplicative ie
y(t) = Level + Trend + Seasonality + Noise 
#### or
 	
y(t) = Level x Trend x Seasonality x Noise

### First Additive Example

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import numpy as np
from random import randrange
import pandas as pd
from matplotlib import pyplot
series = [i+randrange(10) for i in range(1,100)]
result = seasonal_decompose(series, model='additive', freq=1)
result.plot()
pyplot.show()
```

### Multiplicative...


```python
from matplotlib import pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
series = [i**2.0 for i in range(1,100)]
result = seasonal_decompose(series, model='multiplicative', freq=1)
result.plot()
plt.show()
```

### Real Data
Note: the seasonal_decomposition cannot infer the frequency due to how the months are recorded. Hence the pd.to_datetime

```python
import os
series = pd.read_csv(os.getcwd() + '/international-airline-passengers.csv', header = 0)
series.Month = pd.to_datetime(series.Month, errors = 'coerce')
series = series.set_index('Month')
series = series[:-1]
series.plot()
plt.show()
```

```python
result = seasonal_decompose(series, model = 'multiplicative')
result.plot()
plt.show()
```
